```python
import os
import json
import matplotlib.pyplot as plt
```

```python
def plot_data_distribution(data_path, mode):
    name_to_num_images_map = {}
    index_to_name_map = {}
    with open("../Datasets/Kaggle/flower-data/index_to_name.json") as json_file:
        index_to_name_map = json.load(json_file)

    for flower_index in os.listdir(data_path):
        num_images = len(os.listdir(os.path.join(data_path, flower_index)))
        flower_name = index_to_name_map[flower_index]
        name_to_num_images_map[flower_name] = num_images

    plt.figure(figsize = (20, 10))
    plt.bar(name_to_num_images_map.keys(), name_to_num_images_map.values(), align="edge",
            width = 0.4)
    plt.xlabel("Flowers", fontsize=16)
    plt.ylabel("Number of images", fontsize=16)
    plt.title("Distribution of flower images in the " + mode + " set", fontsize=24)
    plt.xticks(rotation="vertical", fontsize=10)
    plt.tight_layout()
    plt.savefig(mode + "_set_distribution.png")
    plt.show()
    print(name_to_num_images_map["sweet william"])
    print(name_to_num_images_map["carnation"])
    print(name_to_num_images_map["english marigold"])
    print(name_to_num_images_map["californian poppy"])
    print(name_to_num_images_map["ball moss"])
    print(name_to_num_images_map["foxglove"])
    print(name_to_num_images_map["camellia"])
    print(name_to_num_images_map["mallow"])
```

```python
training_path = "../Datasets/Kaggle/flower-data/train"
validation_path = "../Datasets/Kaggle/flower-data/val"

plot_data_distribution(training_path, "training")
plot_data_distribution(validation_path, "validation")
```
